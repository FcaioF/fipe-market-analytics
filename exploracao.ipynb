## Retorna marcas motocicletas

```python


# 1. Requisição GET (Buscar dados)
url_get = "https://fipe.parallelum.com.br/api/v2/motorcycles/brands"
response = requests.get(url_get)

if response.status_code == 200:
    #dados = response.json() # Converte resposta para dicionário Python
    for marca in response.json():
        print(marca)
else:
    print(f"Erro no GET: {response.status_code}")
```
